```
import pandas as pd
import numpy as np
df = pd.DataFrame()
```

```
df["Comfort level"] = [12, 28, 28 + 16,28 + 16 + 16,75, 88, 97, 106, 115, 124, 133, 142, 151, 160, 169, 178, 187, 196, 205, 214, 226]
df["Coefficient"] = np.arange(0.0, 10.5, 0.5)
df["Route by foot"] = [34, 31, 31,31, 18, 18, 18,18, 18, 18,18, 18, 18, 18, 18, 18,18, 18, 18,18, 18]
df['Common time'] = [46, 44, 44, 44, 61, 61,61,61,61,61,61,61,61,61,61,61,61,61,61,61, 61]
```

```
import matplotlib.pyplot as plt
plt.plot(df["Coefficient"], df["Comfort level"], 'r-',  linewidth=2, markersize=3, marker='o')
plt.xlabel('Coefficient of preference for walking path')
plt.ylabel('Comfort level')
plt.title('Influence of the coefficient of preference for walking path on the level of comfort')
plt.axhline(y=28, ls='--', c='g')
plt.axhline(y=60, ls='--', c='g')
plt.axhline(y=12, ls='--', c='g')

plt.axvline(0.0, ls='--', c='g')
plt.axvline(0.5, ls='--', c='g')
plt.axvline(1.5, ls='--', c='g')


plt.axhspan(ymin=12, ymax=28, facecolor='b', alpha=0.1)
plt.axhspan(ymin=28, ymax=60, facecolor='g', alpha=0.1)
plt.axhspan(ymin=60, ymax=226, facecolor='m', alpha=0.1)
plt.legend()
plt.show()
#df.plot(x = 'Coefficient', y = 'Route by foot' )
#lines = df[["Common time", "Route by foot"]].plot.line()
```

```
plt.plot(df["Coefficient"], df["Route by foot"], 'r-',  linewidth=2, markersize=3, marker='o', label = 'Wakling part of the route')
plt.plot(df["Coefficient"], df["Common time"], 'b-',  linewidth=2, markersize=3, marker='o', label = 'Common time')
plt.xlabel('Coefficient of preference for walking path')
plt.title('Dependence time from coefficient of preference for walking path')

plt.ylabel('Time')
plt.axvline(0.0, ls='--', c='g')
plt.axvline(0.5, ls='--', c='g')
plt.axvline(1.5, ls='--', c='g')

plt.legend()
plt.show()
```
